```python
import time  
import requests
from bs4 import BeautifulSoup
import pandas as pd

final = pd.DataFrame()

for j in range(68): 
    url = f'https://internshala.com/research-companies/page-{j}/'
    response = requests.get(url)
    
    if response.status_code == 200:
        webpage = response.text
        soup = BeautifulSoup(webpage, 'lxml')
        
        parent_div = soup.find('div', class_='companies_list_container')
        if not parent_div:
            print(f"No data found on page {j}")
            continue  
        
        companies = parent_div.find_all('div', class_='company_card_container')
        
        
        if not companies:
            print(f"No companies found on page {j}")
            continue
        
        name = []
        location = []
        company_size = []
        company_type = []

        for i in companies:
            name.append(i.find('a').text.strip() if i.find('a') else None)
            loc = i.find('div', class_='company_location lower_info')
            location.append(loc.text.strip() if loc else None)
            location = [loc.replace("Location\n", "").strip() for loc in location]
            size = i.find('div', class_='company_size lower_info')
            company_size.append(size.text.strip() if size else None)
            Company_size = [loc.replace("Company size\n", "").strip() for loc in Company_size]
            ctype = i.find('div', class_='company_industry lower_info')
            company_type.append(ctype.text.strip() if ctype else None)
            Company_type = [loc.replace("Industry\n", "").strip() for loc in Company_type]
        
        df = pd.DataFrame({
            'Name': name,
            'Location': location,
            'Company Size': company_size,
            'Company Type': company_type,
        })
        
        
        final = pd.concat([final, df], ignore_index=True)
    else:
        print(f"Failed to fetch page {j}, status code: {response.status_code}")
    
    
    time.sleep(2)

print(final)

```
